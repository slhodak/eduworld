```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import mpld3
```

```python
df = pd.read_csv('../src/data/table-1-1-mf.csv')
grouped = df.set_index('Age Group')
sns.set_theme(style='ticks')
plot = sns.relplot(data=grouped, height=5, aspect=2)
plt.grid(axis='y')
plt.legend()
plt.ylabel('Persons')
figure = plot.fig
```

```python
figure
```

```python
html_str = mpld3.fig_to_html(figure)
Html_file = open("../src/html/attainment.html","w")
Html_file.write(html_str)
Html_file.close()
```
